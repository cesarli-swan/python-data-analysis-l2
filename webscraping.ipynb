```python
import requests, os, bs4
from PIL import Image
import io
url = 'https://xkcd.com'  # the website address
os.makedirs('xkcd', exist_ok=True)  # store comics in a folder called xkcd
counter = 0
max_downlaod = 50
while not url.endswith('#'):
    if counter == max_downlaod:
        break
        
    # TODO: Download the page
    response = requests.get(url)

    # TODO: Find the URL of the comic image
    soup =  bs4.BeautifulSoup(response.text, 'html.parser')
    comic = soup.select('#comic img')

    # TODO: download the image
    comic_url = 'https:' + comic[0].get('src')
    reponse = requests.get(comic_url)
    img = Image.open(io.BytesIO(reponse.content))
        
    # TODO: Save the image to xkcd
    img.save('xkcd/' + os.path.basename(comic_url))
        
    # TODO: Get Prev button's url
    prevLink = soup.select('a[rel="prev"]')[0]
    url = 'https://xkcd.com' + prevLink.get('href')
    counter = counter + 1
print('Done')
```
