```python
iris = datasets.load_iris()
iris_data = iris['data']
```

```python
iris_labels = iris['target']
```

```python
data = pd.DataFrame(iris_data)
labels = pd.DataFrame(iris_labels)
```

```python
df = pd.concat([data, labels], axis=1)
df
```

```python
column_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Species']
column_names
```

```python
column_names[0]
```

```python
df.columns = column_names
df
```

```python
iris_species = iris['target_names']
print(iris_species)
```

```python
species = ['I. setosa', 'I. versicolor', 'I. virginica']
species
```

```python
for index, row in df.iterrows():
    print(row['Species'])
```

```python
species_df = df.copy()
for index, row in df.iterrows():
    species_number = row['Species']

    if species_number == 0.0:
        row['Species'] = species[0]
        
    elif species_number == 1.0:
        row['Species'] = species[1]

    elif species_number == 2.0:
        row['Species'] = species[2]

    species_df.iloc[index] = row

species_df
        
```

```python
ax = species_df.plot(figsize = (12, 6))
ax.axvline(x=0, color='black', linestyle='-')
ax.axvline(x=50, color='black', linestyle='-')
ax.axvline(x=100, color='black', linestyle='-')
ax.axvline(x=150, color='black', linestyle='-')
#ax.axhline(y=4, color='black', linestyle='-')
```

```python
num_bins = len(species_df)
species_df.plot(kind = 'hist', figsize = (12, 6), bins=num_bins)
```

```python
def scatter_species(df, col1, col2):
    df.plot(kind = 'scatter', x=col1, y=col2, c='Species', )
scatter_species(df, 'Sepal Length', 'Sepal Width')
scatter_species(df, 'Petal Length', 'Petal Width')
scatter_species(df, 'Sepal Length', 'Petal Length')
scatter_species(df, 'Sepal Width', 'Petal Width')
scatter_species(df, 'Sepal Width', 'Petal Length')
```
